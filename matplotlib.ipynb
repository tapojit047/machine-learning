```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0,5,11)
y = x**2

x
```

```python
y
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,'r')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Experiment")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'b--')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y , 'g')

axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Title")
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```
